```python
from lib import Compressors
import torch
import matplotlib.pyplot as plt
import tqdm

def hilbert_matrix(n, m):
    i = torch.arange(n, dtype=torch.float64).reshape(-1, 1)
    j = torch.arange(m, dtype=torch.float64).reshape(1, -1)
    H = 1.0 / (i + j + 1)
    return H.float()
```

```python
m = 500
sizes = [16, 50, 100, 500, 1000, 1500, 2000]
r = 16
k = 10
compress = [Compressors.PISSA, Compressors.COTAN_HALF, Compressors.COTAN, Compressors.CORDA]

def clc(m, n, distribution, k):
    data = [0] * len(compress)
    for _ in range(k):
        W = distribution.sample((m, n)).to('cuda')
        X = torch.randn((n, 10000)).to('cuda')
        # X = distribution.sample((n, 10000)).to('cuda')
        # X = hilbert_matrix(n, 10000).to('cuda')
        
        for i in range(len(compress)):
            if i:
                A, B = compress[i](W, X, r)
            else:
                A, B = compress[i](W, r)
            T = A @ B
            data[i] += torch.norm(T) / k

    return data

distribution = torch.distributions.Exponential(rate=500)
# distribution = torch.distributions.Normal(loc=0.0, scale=1.0)
# distribution = distribution = torch.distributions.Uniform(low=-1.0, high=1.0)


results = []

for size in tqdm.tqdm(sizes):
    data = clc(m, size, distribution, k)
    results.append(data)

# Plotting the results
plt.figure(figsize=(10, 6))

for i in range(len(compress)):
    compressor_name = compress[i].__name__
    plt.plot(sizes, [res[i].cpu() for res in results], label=f'{compressor_name}')

plt.xlabel('Matrix Size (n)')
plt.ylabel('Norm of T')
plt.title('Comparison of Compressors')
plt.legend()
plt.grid(True)
plt.show()
```
